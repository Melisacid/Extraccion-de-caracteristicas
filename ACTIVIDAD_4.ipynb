## Aplicar métodos (Desviación Estándar o Rango Intercuartílico) para eliminar Outliers del dataframe.

**Rango Intercuartílico** puesto que este método es mas robusto, evitando el problema que se presenta con la información puesto que la distribucion de datos es anormal y con valores atípicos que impacten la media y desviación. 

```python
#SEPARAMOS LAS COLUMNAS CUANTITATIVAS DEL DATA2
import pandas as pd
# Verificar los tipos de datos de cada columna
print(Micro_Retailer2.dtypes)
# Seleccionar las variables numéricas
data_num = Micro_Retailer2.select_dtypes(include="number")
# Mostrar el DataFrame resultante
print(data_num)
```

```python
data_num.info()
#nos damos cuenta que son numericas por el dtype 
#float64 e int64
```

```python
data_num.head()
```

```python
data_num2=data_num.copy()
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data_num2.plot(kind='box', vert=False)
plt.title("Valores Atípicos de Girona_España")
plt.show() #dibujamos el diagrama
```

```python
#Encuentro cuartiles 0.25 y 0.75
y=data_num2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
```

```python
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers= data_num2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers
```

```python
#Obtenemos datos limpios del Dataframe
data3= data_num2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4=data3.copy()
data4=data4.fillna(round(data3.mean(),1))
data4
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(15, 8))
plt.boxplot(data4) 
plt.title("Outliers de Girona_España")
plt.show() #dibujamos el diagrama
```
